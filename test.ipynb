```python
def CalcSalary(d1:dict,d2:dict,d3:dict):
    print("TechCorp")
    
    for j in d1:
        if j['contract_hours'] < j['weekly_hours_worked']:
            sup_hour = j['weekly_hours_worked']-j['contract_hours']
            salary = (j['hourly_rate']*j['contract_hours'] + sup_hour * j['hourly_rate']*1.5)*4

        else:
            salary = j['hourly_rate']*j['weekly_hours_worked']*4

        
        print(f"{j['name']} | {j['job']} | Salaire mensuel : {salary:.2f} €")

    for k in d2:
        if k['contract_hours'] < k['weekly_hours_worked']:
            sup_hour = k['weekly_hours_worked']-k['contract_hours']
            salary = (k['hourly_rate']*k['contract_hours'] + sup_hour * k['hourly_rate']*1.5)*4

        else: 
            salary = k['hourly_rate']*k['weekly_hours_worked']*4

        print(f"{k['name']} | {k['job']} | Salaire mensuel : {salary:.2f} €")

    for l in d3:
        if l['contract_hours'] < l['weekly_hours_worked']:
            sup_hour = l['weekly_hours_worked']-l['contract_hours']
            salary = (l['hourly_rate']*l['contract_hours'] + sup_hour * l['hourly_rate']*1.5)*4

        else: 
            salary = l['hourly_rate']*l['weekly_hours_worked']*4

        print(f"{l['name']:<10}| {l['job']:<10}| Salaire mensuel : {salary:.2f<10} €")
```
